# Linearly separable dataset

```python
data1 = np.random.rand(200, 3) * 0.5
data1[:, 0] *= data1[:, 1]
data1[:, 2] = np.zeros(200)
data1 = data1[data1[:, 1] > 0.05]

data2 = np.random.rand(200, 3) * 0.5
data2[:, 1] *= data2[:, 0]
data2[:, 2] = np.ones(200)
data2 = data2[data2[:, 0] > 0.05]

data = np.vstack((data1, data2))
target = data[:, -1]
data = data[:, :2]

fig = plt.figure(figsize=(8, 8))
ax = plt.axes()
ax.scatter(data[:, 0], data[:, 1], c=target, cmap="bwr")
```

# Iris dataset

```python
import plotly.graph_objects as go

iris = datasets.load_iris(as_frame=False)
dataset = np.hstack((iris.data, np.reshape(iris.target, (-1, 1))))
dataset = dataset[dataset[:, -1] != 2]

target = dataset[:, -1]
data = dataset[:, :-1]

fig = go.Figure(data=[go.Scatter3d(x=data[:, 0], 
                                   y=data[:, 1], 
                                   z=data[:, 2], 
                                   mode="markers",
                                   marker=dict(
                                       size=4,
                                       color=target,
                                       colorscale='Viridis',
                                       opacity=0.8)
                                  )
                     ]
               )
fig.show()

# If plotly isn't default for colab notebooks. Note you can only plot figures inline in Google colab.
#%matplotlib qt
#fig = plt.figure(figsize=(8, 8))
#ax = plt.axes(projection='3d')
#ax.scatter3D(data[:, 0], data[:, 1], data[:, 2], c=target)
```
